```python
import pandas as pd

# Load your data (replace with your path)
df = pd.read_csv("labels_training.tsv", sep='\t')

# Identify value columns
attained_cols = [col for col in df.columns if 'attained' in col.lower()]
constrained_cols = [col for col in df.columns if 'constrained' in col.lower()]

# Ensure all value columns are numeric (important!)
df[attained_cols + constrained_cols] = df[attained_cols + constrained_cols].apply(pd.to_numeric, errors='coerce').fillna(0)

# Classify each row based on the logic
def classify(row):
    attained_sum = row[attained_cols].sum()
    constrained_sum = row[constrained_cols].sum()
    
    if attained_sum > 0 and constrained_sum == 0:
        return 'Attained'
    elif constrained_sum > 0 and attained_sum == 0:
        return 'Constrained'
    elif attained_sum > 0 and constrained_sum > 0:
        return 'Mixed'
    else:
        return 'None'

df['Value_Reference_Type'] = df.apply(classify, axis=1)

# Count the classification
counts = df['Value_Reference_Type'].value_counts()
print(counts)
```
